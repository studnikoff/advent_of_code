# Part 1


```python
with open('advent_input2.txt', 'r') as f:
    inp = f.read()
```

```python
inp = [i.split(' ') for i in inp.split('\n')]
```

```python
inp
```

```python
from cyclic_group import CyclicGroup, RPS
```

```python
mapper = {
    'X': 'A',
    'Y': 'B',
    'Z': 'C'
}
rps = RPS(('B', 'A', 'C'), mapper)
```

```python
rps_score = {
    'X': 1,
    'Y': 2,
    'Z': 3
}
sum_res = 0
for round in inp[:-1]:
    a = rps_score[round[1]]
    b = rps.round_score(round[0], round[1])
    sum_res = sum_res+a+b
```

```python
sum_res
```

# Part 2


```python
rps_score = {
    'A': 1,
    'B': 2,
    'C': 3
}

cg = CyclicGroup(('B', 'A', 'C'))
```

```python
sum_res = 0
for round in inp[:-1]:
    if round[1] == 'X':
        value = cg.next(round[0])
        sum_res = sum_res + 0 + rps_score[value]
    elif round[1] == 'Y':
        value = round[0]
        sum_res = sum_res + 3 + rps_score[value]
    else:
        value = cg.previous(round[0])
        sum_res = sum_res + 6 + rps_score[value]
print(sum_res)
```
