```python
# Prepare table with score values

import copy
lowercase = list(map(chr, range(97,123)))
uppercase = [i.capitalize() for i in lowercase]

lowercase.extend(uppercase)
items = copy.deepcopy(lowercase)
del lowercase
del uppercase

tablescore = {items[i-1]:i for i in range(1, 53)}
del items
```

```python
print(tablescore)
```

```python
# rucksacks = test
rucksacks = inp.split('\n')
score = 0
for ruck in rucksacks:
    N = len(ruck)
    mid = N//2
    hashset = set(ruck[:mid])
    for i in ruck[mid:]:
        if i in hashset:
            item = i
            score += tablescore[i]
            break
    print(ruck, item)

print(score)
```

# Part 2


```python
rucksacks = inp.split('\n')
groups = [rucksacks[x:x+3] for x in range(0, len(rucksacks), 3)]
```

```python
score = 0
for group in groups:
    sets = list(map(set, group))
    res = sets[0] & sets[1] & sets[2] # as only 1 elf has always 1 badge
    badge = str(res.pop())
    score += tablescore[badge]
print(score)
```
